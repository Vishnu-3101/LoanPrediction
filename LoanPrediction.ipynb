```python
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['train_ctrUa4K (1).csv']))
```

```python
df.drop('Loan_ID',axis = 1,inplace = True)
```

```python
df.head()
```

```python
df2 = df.copy()
```

```python
df2['Property_Area'].value_counts()
```

```python
df3 = pd.concat([pd.get_dummies(df2['Gender']),df2],axis= 1)
```

```python
gen_dum = pd.get_dummies(df2['Gender'])
mar_dum = pd.get_dummies(df2['Married'])
edu_dum = pd.get_dummies(df2['Education'])
sel_dum = pd.get_dummies(df2['Self_Employed'])
pro_dum = pd.get_dummies(df2['Property_Area'])
```

```python
df4 = pd.concat([gen_dum,mar_dum,edu_dum,sel_dum,pro_dum,df2],axis= 1)
```

```python
df4.drop(['Gender','Married','Education','Self_Employed','Property_Area'],axis=1,inplace=True)
```

```python
dep_dum = pd.get_dummies(df2['Dependents'])
df5 = pd.concat([dep_dum,df4],axis =1)
```

```python
df5.drop(['3+','Male','Yes','Graduate','Urban','Dependents'],axis=1,inplace=True)
```

```python
df5.head()
```

```python
df5['LoanAmount'].fillna(df5['LoanAmount'].mean(),inplace=True)
```

```python
df5['Loan_Amount_Term'].fillna(df5['Loan_Amount_Term'].mean(),inplace=True)
df5['Credit_History'].fillna(df5['Credit_History'].mean(),inplace=True)
```

```python
df5.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
cols = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']
df5[cols]=scaler.fit_transform(df5[cols])
```

```python
df5.head()
```

```python
X = df5.drop('Loan_Status',axis=1)
```

```python
X.head()
```

```python
Y = df5['Loan_Status']
```

```python
Y.head()
```

```python
Y = pd.get_dummies(Y).drop('N',axis =1)
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,stratify = Y)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```
